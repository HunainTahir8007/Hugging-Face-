```python
from transformers import AutoModelForSequenceClassification
from transformers import AutoTokenizer
import torch
```

```python
class Manualmodel:
  def __init__(self):
    print("Loading model ....")
    model_name = "distilbert-base-uncased-finetuned-sst-2-english"
    self.tokenizer = AutoTokenizer.from_pretrained(model_name)
    self.model = AutoModelForSequenceClassification.from_pretrained(model_name)
    self.id2label = self.model.config.id2label
    print("Model loaded ✅\n")

  def tokenize(self,text):
    return self.tokenizer(text, return_tensors="pt", padding = True , truncation = True )

  def predict(self,text):
    inputs = self.tokenize(text)
    with torch.no_grad():
      outputs = self.model(**inputs)
      probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)
      predicted_class_idx = torch.argmax(probabilities, dim=-1).item()
      label = self.id2label[predicted_class_idx]
      confidence = probabilities[0][predicted_class_idx].item()
      return label,confidence

  def final_out(self,text):
        label,confidence = self.predict(text)
        print(f"Text:       {text}")
        print(f"Prediction: {label}")
        print(f"Confidence: {confidence*100:.2f}%")
        print("─" * 50)
        return label,confidence

```

```python
engine = Manualmodel()
texts = [
    "I built a Transformer from scratch!",
    "Financial problems are really stressful",
    "HuggingFace is going to get me my first dollar",
    "University and freelancing together is hard",
    "I will never give up on my goals"
]

for text in texts:
    res=engine.predict(text)
```
